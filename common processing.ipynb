```python
import pandas as pd
data = pd.read_csv('hourly_data_with_consent.csv')
```

```python
vedanta_data = data[data['plant_name'] == 'Vedanta Ltd.']
```

```python
missing_emission_vedanta_data = vedanta_data[vedanta_data['revised_emission'] == 999]
```

```python
vedanta_data.info()
    
```

```python
vedanta_data = vedanta_data[['plant_name','CEMS','timestamp','emission','modified consent']]
```

```python
vedanta_data.head()
```

```python
vedanta_data['revised_emission'] = vedanta_data['emission']
```

```python
vedanta_data['revised_emission'].fillna(0, inplace=True)
```

```python
miss_data = vedanta_data[vedanta_data['revised_emission'] == 0]
miss_data
```

```python
vedanta_data['exceedance'] = vedanta_data['revised_emission'] - vedanta_data['modified consent']
```

```python
def fun(x):
    if x < 0:
        y = 0
    else:
        y = x
    return y
```

```python
vedanta_data['revised_exceedance'] = vedanta_data['exceedance'].apply(lambda x : fun(x))
```

```python
pc0_100 = sum(vedanta_data['revised_exceedance'] * 0.09 * 100)
```

```python
_ = plt.figure(figsize=(12,8))
_ = plt.xlabel('rate per Kg', fontsize = 14)
_ = plt.ylabel('pollution charge in INR', fontsize = 14)
_ = plt.plot([100,500,1000],[pc0_100,pc0_500,pc0_1000],'go--', label='proxy=0')
_ = plt.plot([100,500,1000],[pc200_100,pc200_500,pc200_1000],'ro--', label = 'proxy=100')
_ = plt.plot([100,500,1000],[pc999_100,pc999_500,pc999_1000],'bo--', label = 'proxy=999')
_ = plt.title('Plotting pollution charge as a function of different rate and proxy emission for CEMS1 Vedanta', fontsize = 16)
_ = plt.legend()
plt.show()
```
