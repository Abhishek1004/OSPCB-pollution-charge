```python
from datetime import date, datetime, timedelta

def datetime_range(start, end, delta):
    current = start
    if not isinstance(delta, timedelta):
        delta = timedelta(**delta)
    while current < end:
        yield current
        current += delta

li=[]
start = datetime(2018,6,1)
end = datetime(2018,7,1)

for dt in datetime_range(start, end, {'days': 0, 'hours':1}):
     li.append(dt.strftime('%m/%d/%Y %H:%M:%S'))
```

```python
plant_names = data.plant_name.unique()
```

```python
len(plant_names)
```

```python
data_groupedby_hour['timestamp'] = pd.to_datetime( data_groupedby_hour['timestamp'], format = '%m/%d/%Y %H:%M:%S' ) 
```

```python
converted_data = pd.DataFrame(columns=['plant_name','CEMS','pollutant','timestamp','emission'])
for plant_name in plant_names:
    plant_data = data_groupedby_hour[data_groupedby_hour['plant_name'] == plant_name]
    CEMS_names = plant_data.CEMS.unique()
    for CEMS in CEMS_names:
        CEMS_data = plant_data[plant_data['CEMS'] == CEMS]
        new_timestamp_data = timestamp_data.copy()
        new_timestamp_data['plant_name'] = plant_name
        new_timestamp_data['CEMS'] = CEMS
        new_timestamp_data['pollutant'] = 'PM'
        new_data = pd.merge(new_timestamp_data, CEMS_data, how='left', on=['plant_name', 'CEMS', 'pollutant','timestamp'])
        new_data = new_data[['plant_name','CEMS','pollutant','timestamp','emission']]
        frames = [converted_data, new_data]
        converted_data = pd.concat(frames)
        
        
```

```python
len(converted_data)
```

```python
standard_data = pd.read_excel('plant_station_with_consent_8jun2018.xlsx')
```

```python
hourly_data_with_consent = pd.merge(converted_data, standard_data, how = 'left', on = ['plant_name', 'CEMS'])
```

```python
hourly_data_with_consent.tail(500)
```

```python
hourly_data_with_consent['modified consent'].fillna(100, inplace = True)
```

```python
hourly_data_with_consent['emission_exceedance'] = hourly_data_with_consent['revised_emission'] - hourly_data_with_consent['modified consent']
```

```python
hourly_data_with_consent['emission'].isnull().values.ravel().sum()
```

```python
hourly_data_with_consent['exceedance'] = hourly_data_with_consent['revised_emission'] - hourly_data_with_consent['modified consent']
```

```python
hourly_data_with_consent.head(100)
```

```python
def fun(x):
    if x < 0:
        y = 0
    else:
        y = x
    return y
```

```python
hourly_data_with_consent['revised_exceedance'] = hourly_data_with_consent['exceedance'].apply(lambda x : fun(x))
```

```python
hourly_data_with_consent.head(100)
```

```python
hourly_data_with_consent['pollution_charge'] = hourly_data_with_consent['revised_exceedance'] * 0.09 * 100
```

```python
hourly_data_with_consent.head(100)
```

```python
hourly_data_with_consent.to_csv('hourly_data_with_consent.csv')
```

```python
hourly_data_with_consent_groupby_plant_CEMS = hourly_data_with_consent.groupby(['plant_name','CEMS'])['pollution_charge'].sum()
```

```python
hourly_data_with_consent_groupby_plant_CEMS = hourly_data_with_consent_groupby_plant_CEMS.reset_index()
```
